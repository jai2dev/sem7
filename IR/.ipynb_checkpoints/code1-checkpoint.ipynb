```python

file = open("news.txt", "r")

vocab = []
for line in file:

  for word in line.split():

      vocab.append(word)

# print(vocab)
initial_vocab_size = len(vocab)
print('initial vocabulary size: ',len(vocab))
```

```python
f = open('news.txt', encoding='utf8')
read = f.read()
```

```python
# read = file.read();
file.seek(0)

line = 1
for word in read:
    if word == '\n':
        line += 1
print("Number of lines in file is: ", line)
  
# create a list to
# store each line as
# an element of list
array = []
for i in range(line):
    array.append(file.readline())
  
array
```

```python
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
for ele in read:  
    if ele in punc:  
        read = read.replace(ele, " ")        
read
```

```python
# to maintain uniformity
read=read.lower()                    
read
```

```python
from nltk.tokenize import word_tokenize
# from nltk.corpus import stopwords
# nltk.download('stopwords')
  
for i in range(1):
    # this will convert
    # the word into tokens
    text_tokens = word_tokenize(read)
tokenize_vocab_size = len(text_tokens)
print(len(text_tokens))
print(text_tokens)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
tokens_without_sw = [
    word for word in text_tokens if not word in stopwords.words()]
stop_word_vocab_size = len(tokens_without_sw)
print(len(tokens_without_sw))
print(tokens_without_sw)
```

```python
#stemming
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens_without_sw]
print(stemmed_tokens)
```

```python
print('Vocabulary Sizes after various preprocessing steps')
print('Initial vocabulary size: ',initial_vocab_size)
print('After tokenization: ',tokenize_vocab_size)
print('After removing stop words: ',stop_word_vocab_size)
print('After Stemming',len(stemmed_tokens))
```

```python
#functions for performing AND,OR, and NOT on two sets
def do_and(set1,set2):
  result = set1.intersection(set2)
  return result
def do_or(set1,set2):
  result = set1.union(set2)
  return result
def do_not(set1,set2):
  result = set1-set2
  return result
```

```python
def search_doc(query):
  print("Query:",query)
  final_doc_set = set()
  prev = 'OR'
  for word in query.split():
    if word not in ["AND","OR","NOT"]:
      print(word,dict[word])
      cur_posting = dict[word][1] 
      if prev=='AND':
        # print('AND called')
        final_doc_set = do_and(final_doc_set,cur_posting)
      elif prev=='OR':
        # print('OR called')
        final_doc_set = do_or(final_doc_set,cur_posting)
      elif prev=='NOT':
        # print('NOT called')
        final_doc_set = do_not(final_doc_set,cur_posting)
    else:
      prev = word

  
  if len(final_doc_set)==0:
    print('Result Set:','Empty Set')
  else:
    print('Result Set:',final_doc_set)
  print('\n')
```

```python
#performing queries 
query1 = "bharat OR delta OR low NOT given"
search_doc(query1)
```

```python
query2 = "third AND pandem AND low"
search_doc(query2)
```
