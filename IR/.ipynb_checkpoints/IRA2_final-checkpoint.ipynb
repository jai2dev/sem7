```python
def BOW(path):
    print(path)
    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    
    terms = []
    
    file = open(path, 'r')
    text = file.read()
    
    bow = text.split(" ")
     
                
    for word in bow:
        if '\n' in word:
            word = word.replace('\n', "")
        for char in word:
            if char in punc:
                #print("replaced", word)
                word = word.replace(char, "")
        
        terms.append(word.lower())
        
        
    return terms
            
```

```python
BOW('IR A2 Corpus/doc1')
```

```python
wordSet = set()
basepath = 'IR A2 Corpus/'
for entry in os.listdir(basepath):
    path = os.path.join(basepath, entry)
    res = BOW(path)
    
    for word in res:
        wordSet.add(word)
        
print(wordSet)
    
```

```python
import math
```

```python
def TF(wordDict):
    
    max_fi = 0
    for key, val in wordDict.items():
        if val>max_fi:
            max_fi = val
    
    newDict = {}
    for key, val in wordDict.items():
        if val>0:
            
            #log normalization
            #newDict[key] = 1+ math.log(val,2)
            
            
            #double normalization K
            newDict[key] = 0.5 * ((0.5*val)/ max_fi)
    
    return newDict
```

```python
dictList = [dict() for x in range(N)]
for i in range(N):
    dictList[i] = dict.fromkeys(wordSet, 0) 
    
    for word in BOW('IR A2 Corpus/doc'+str(i)):
        dictList[i][word]+=1
```

```python
dictList[0]
```

```python
def computeIDF(dictList):
    idfDict = {}
    N = len(dictList)
    print(N)
    
    idfDict = dict.fromkeys(dictList[0].keys(), 0)
    
    for doc in dictList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        #inverse frequency
        #idfDict[word] = math.log((N / float(val)), 2)
        
        #inverse frequency smooth
        idfDict[word] = math.log((1+(N/float(val))), 2)
        
        #probabilistic inverse frequency
        #idfDict[word] = math.log(((N-val)/val), 2)
        
    return idfDict
```

```python
idfs = computeIDF(dictList)
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
finalTF = [dict() for x in range(N)]
finalTFIDF = [dict() for x in range(N)]
for i in range(N):
    finalTF[i] = TF(dictList[i])
```

```python
for i in range(N):
    finalTFIDF[i] = computeTFIDF(finalTF[i], idfs)
```

```python
query = "the men China"

queryTerms = query.split(' ')
for i in range(len(queryTerms)):
    queryTerms[i] = queryTerms[i].lower()
    
queryTerms   
```

```python
queryCount = dict.fromkeys(queryTerms, 0) 
for term in queryTerms:
    queryCount[term]+=1
    
queryCount
```

```python
for word, val in queryCount.items():
        queryCount[word] = 1 + math.log(val, 2)
queryCount
```

```python
queryTFIDF = dict.fromkeys(queryTerms, 0) 
for word in queryTerms:
    print(word)
    queryTFIDF[word]=queryCount[word]*idfs[word]

queryTFIDF
```

```python
q_square_sum = 0 

for word in queryTerms:
    q_square_sum += queryTFIDF[word]**2
    

q_square_sum = math.sqrt(q_square_sum)
```

```python
doc_square_sum = [0]*N

for i in range(N):
    for word in wordSet:
        doc_square_sum[i] += (df.at[i, word]**2) 
    
    doc_square_sum[i] = math.sqrt(doc_square_sum[i])
    
```

```python
cos_sim = [0]*N
 
```

```python
for i in range(N):
    num = 0
    den = doc_square_sum[i]*q_square_sum

    for word in queryTerms:
        table_val = df.at[i, word]
        num += queryTFIDF[word]*table_val
        cos_sim[i]= num/den

    
```
