```python
def preprocess(read):
    
    #Step 1: Removal of Punctuation
    punctuation = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
    for element in read:  
        if element in punctuation:  
            read = read.replace(element, " ")        
    
    #Step 2: Changing case to lower case 
    read=read.lower()
    text_tokens = word_tokenize(read)
    tokenize_vocab_size = len(text_tokens)
    
    #Step 3: Removing Stop words
    tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
    stop_word_vocab_size = len(tokens_without_sw)
    tokens_without_sw
    
    #Step 4: Stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word) for word in tokens_without_sw]
    print(stemmed_tokens)
    return stemmed_tokens
```

```python
#Preprocess each doc and making tokens for each doc
docs_data=[]
for i in range(len(array)):
    x=preprocess(array[i])
    docs_data.append(x)  
print(docs_data)
```

```python
#Calculating Df values for each term in vocabulary
DF={}
for i in range(len(docs_data)):
    token=docs_data[i]
    for word in token:
        try:
            DF[word].add(i)
        except:
            DF[word]={i}
for i in DF:
    DF[i]=len(DF[i])
```

```python
print(DF)
```

```python
vocabulary=[]
for i in DF:
    vocabulary.append(i)
print(len(vocabulary))
print(vocabulary)
```

```python
#Calculating tf-idf values for each term in the vocabulary
cnt=0
tf_idf={}
for i in range(len(docs_data)):
    token=docs_data[i]
    words=len(token)
    count=Counter(token)

    for x in np.unique(token):
            tf=count[x]/words
            df=DF[x] if x in vocabulary else 0
            idf=np.log(len(vocabulary)+1/df+1)
            tf_idf[cnt,x]=tf*idf
    cnt+=1
```

```python
tf_idf
```

```python
# Table creation 
Table=np.zeros((cnt,len(vocabulary)))
for i in tf_idf:
    idx=vocabulary.index(i[1])
    Table[i[0]][idx]=tf_idf[i]
Table
```

```python
def make_vector(x):
    array = np.zeros((len(vocabulary)))
    
    count = Counter(x)
    words = len(x)

    query_weights = {}
    
    for c in np.unique(x):
        
        tf = count[c]/words
        df = DF[c] if c in vocabulary else 0
        idf = np.log((len(vocabulary)+1)/(df+1))

        try:
            idx = vocabulary.index(c)
            array[idx] = tf*idf
        except:
            pass
    return array
```

```python
# Considering cosine similarity for closeness measure
def cosine_score(a, b):
    cos_score = np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))
    return cos_score
```

```python
def cosine_similarity(k, query,table):
    tokens = query.split()
    d_cosines = []
    query_vector = make_vector(tokens)
    for d in table:
        d_cosines.append(cosine_score(query_vector, d))
    if k == 0:
        out = np.array(d_cosines).argsort()[::-1]
    else:
        out = np.array(d_cosines).argsort()[-k:][::-1]
    return out
```
